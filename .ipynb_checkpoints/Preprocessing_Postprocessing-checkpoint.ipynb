## Preprocessing

During this step the images should be:
- Resized to the same resolution (we used 224 Ã— 224).
- If training will be performed, normalized to [0, 1].
- Split into a training and a test set. 

### Resize Images

```python
from PIL import Image
import numpy as np
import glob
import os
import cv2
```

```python
translate = {"n02085620": "Chihuahua",
             "n02085782": "Japanese_spaniel",
             "n02085936": "Maltese_dog",
             "n02086079": "Pekinese",
             "n02086240": "Shih-Tzu"
             }
```

```python
#resize & normalize & save
raw="data\\raw" #the path to raw images
path = 'data\\resized\\test'    #the path where to save resized images 
                                #If images are used for training,path = 'data\\resized'

for k,v in translate.items():
    if not os.path.exists(path+"\\"+k):
        os.makedirs(path+"\\"+k) 
    for FILENAME in glob.glob(raw+'\\'+k+"-"+v+"\\*.jpg"): #path of raw images
        image = cv2.imread(FILENAME, cv2.IMREAD_COLOR)  # uint8 image
        img = cv2.resize(image,(224,224))
        #img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        new_name=os.path.split(FILENAME)[1].replace("jpg", "jpeg")
        new_path=path+'\\'+ k + "\\" + new_name
        cv2.imwrite(new_path,img)
```

#### If images are used for training, below code chunk can be used to split images into train/test sub directories

```python
#Split main folder with category subfolders into train & test folders with category subfolders
import splitfolders
input_folder=path
outputdir="data\\img"
splitfolders.fixed(input_folder, output=outputdir, seed=1337, fixed=(100), oversample=False, group_prefix=None)
```
